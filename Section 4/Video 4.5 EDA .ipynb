```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
```

```python
def myplot(data,playerlist=Players):
    Col = {"DeMarDeRozan":"Black","StephenCurry":"red","LeBronJames":"Blue","NikolaJokic":"Green","BogdanBogdanovic":"Yellow"}
    Mrk = {"DeMarDeRozan":"s","StephenCurry":"o","LeBronJames":"D","NikolaJokic":"^","BogdanBogdanovic":"D"}
    for name in playerlist:
        plt.plot(data[Pdict[name]], c= Col[name], ls='--', marker= Mrk[name], ms=7, label = name)
        plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,5)),Seasons, rotation = 'vertical')
    plt.show()
```

```python
myplot(Salary)
```
